```python
import pandas as pd
import os

file_path = r'C:\Users\iamke\OneDrive\Desktop\Methodology\DATASET.csv'

print(f"Attempting to read file at: {file_path}")

if not os.path.exists(file_path):
    print(f"File not found at path: {file_path}")
else:
    print("File exists.")
    try:
        print("Loading the CSV file into a DataFrame...")
        df = pd.read_csv(file_path)
        print("CSV file loaded successfully.")
        
        if df.empty:
            print("CSV file is empty. Please check if the file contains data.")
        else:
           
            print("First 10 Rows of the CSV file:")
            print(df.head(10)) 
        print("\nDataFrame Columns:")
        print(df.columns)
    
    except pd.errors.EmptyDataError:
        print("CSV file is empty. Please check if the file contains data.")
    except pd.errors.ParserError:
        print("There was a parsing error while reading the CSV file. Please check if the file is in the correct format.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python

positive_results_df = df[df['result'] == 1]
            
print("\nRespondents with positive results (result = 1):")
print(positive_results_df)
 
```

```python
negative_results_df = df[df['result'] == -1]
            
print("\nRespondents with negative results (result = -1):")
print(negative_results_df)
```

```python
neutral_results_df = df[df['result'] == 0]
            
print("\nRespondents with neutral results (result = 0):")
print(neutral_results_df)
```

```python
import csv


positive_count = 0
negative_count = 0
neutral_count = 0

with open(r'C:\Users\iamke\OneDrive\Desktop\Methodology\DATASET.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    
    for row in reader:
        
        result = int(row['result']) 
        if result > 0:
            positive_count += 1
        elif result < 0:
            negative_count += 1
        else:
            neutral_count += 1

print("Positive count:", positive_count)
print("Negative count:", negative_count)
print("Neutral count:", neutral_count)
```
